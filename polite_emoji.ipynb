```python
emo = []
for i in range(4):
    for row in tables[i].find_all('tr')[1:]:
        for td in row.find_all('td')[:-2]:
            s = str(td)
            if re.match(r".+<br/>.+",s):
                m = re.findall(r"<td>(.+)?<br/>(.+)?</td>",s)
                try:
                    emo += m[0]
                except:
                    continue
            else:
                emo.append(td.text)

final = []
for e in emo:
    if e.strip():
        final += e.strip().split(" ")
```

```python
emo = []
for i in [4, 5]:
    for row in tables[i].find_all('tr')[1:]:
        #print(row)
        for td in row.find_all('td')[:-2]:
            s = str(td)
            #print(s, "\n")
            if re.match(r"<td><span style=\"white-space:nowrap\">.+",s):
                m = re.findall(r"<td><span style=\"white-space:nowrap\">(.+)</span></span></td>",s)
                try:
                    new = m[0].replace("<span style=\"margin-left:1.5em\">", 'ม')
                    new = new.replace("</span><span style=\"margin-left:1.5em\">", 'ม')
                    new = new.replace("</span></span> <span style=\"white-space:nowrap\">", 'ม')
                    new = new.replace("</span>", '')
                    emo += new.split("ม")
                except:
                    continue
            else:
                emo.append(td.text)
final += emo
```

```python
import unicodedata
def extract_emojis(tw, emoticon = False):
    '''
    extracting emoji name from string
    
    input: (string) a tweet
    output: (dictionary) with key emoji storing a list of
                emoji names and key text storing the text
                without emoji
    '''
    tw = str(tw)
    d = {}
    d['emoji'] = []
    text = ''
    d['text'] = ''
    
    # detecting standard emoji 
    for c in tw:
        if c in UNICODE_EMOJI:
            try:
                if (not 'EMOJI MODIFIER' in unicodedata.name(c)) & \
                    (not 'REGIONAL INDICATOR' in unicodedata.name(c)):
                    d['emoji'].append(unicodedata.name(c))
            except:
                continue
        else:
            text += c
    
    # detecting Tsuki's special emoji and direct mention
    for word in text.split():
        # stand alone emoji
        if re.match(r"<:[a-zA-Z]+:[0-9]+>",word):
            d['emoji'].append(re.findall(r"<.?:([a-zA-Z]+):[0-9]+>",word)[0])
        # emoji and text without space
        elif re.match(r"(.+)(<:[a-zA-Z]+:[0-9]+>)",word):
            d['emoji'].append(re.match(r"(.+)(<:[a-zA-Z]+:[0-9]+>)",word).group(2))
            d['text'] += re.match(r"(.+)(<:[a-zA-Z]+:[0-9]+>)",word).group(1) + " "
        # direct mention
        elif re.match(r"<@[0-9!]+",word):
            d['text'] += '<user_mention>' + " "
        # word then direct mention without space
        elif re.match(r"(.+)(<@[0-9!]+)",word):
            d['text'] += re.match(r"(.+)<@[0-9!]+",word).group(1) + '<user_mention>' + " "
        # direct mention then word without space
        elif re.match(r"(<@[0-9!]+)(.+)",word):
            d['text'] += '<user_mention>' + re.match(r"(.+)<@[0-9!]+",word).group(1) + " "
        # if emoticons are to be counted as emoji
        elif emoticon == True and word in final:
            d['emoji'].append(word)
        # normal word
        else:
            d['text'] += word + " "
    return d

def get_emoji(d):
    '''extracting list of emojis from dictionary'''
    return d['emoji']

def get_text(d):
    '''extracting text from dictionary'''
    return d['text']

def is_emoji(lst):
    return len(lst) != 0

def num_emoji(lst):
    return len(lst)
```

```python
tsuki['dict'] = tsuki['content'].apply(extract_emojis)
tsuki['emoji'] = tsuki['dict'].apply(get_emoji)
tsuki['text'] = tsuki['dict'].apply(get_text)
tsuki['is_emoji'] = tsuki['emoji'].apply(is_emoji)
tsuki['num_emoji'] = tsuki['emoji'].apply(num_emoji)

tsuki.iloc[::500]
```

```python
tsuki_final = tsuki[tsuki["text"] != ""]
tsuki_final.shape
```

```python
# data for polite level classification
wiki = pd.read_csv("wikipedia.annotated.csv")
wiki = wiki[['Request', 'Normalized Score']]
stack = pd.read_csv("stack-exchange.annotated.csv")
stack = stack[['Request', 'Normalized Score']]
corpus = stack.append(wiki)

# binning polite score into 5 equal-interval bins
two = corpus['Normalized Score'].quantile(.2)
four = corpus['Normalized Score'].quantile(.4)
six = corpus['Normalized Score'].quantile(.6)
eight = corpus['Normalized Score'].quantile(.8)

def classify(score):
    if score < two:
        return "0"
    elif score < four:
        return "1"
    if score < six:
        return "2"
    elif score < eight:
        return "3"
    else:
        return "4"

corpus['Polite'] = corpus['Normalized Score'].apply(classify)
```

```python
corpus.index = range(0, corpus.shape[0])
```

```python
# preparing data to train a max entropy classifyer
def dialogue_act_features(post):
    features = {}
    for word in nltk.word_tokenize(post):
        features['contains({})'.format(word.lower())] = True
    return features

featuresets = [(dialogue_act_features(post['Request']), post['Polite'])
              for post in corpus.T.to_dict().values()]

# 80/20 split
size = int(len(featuresets) * 0.2)
train_set, test_set = featuresets[size:], featuresets[:size]
```

```python
# training classifier and check accuracy on the test set
classifier = nltk.MaxentClassifier.train(featuresets)
print(nltk.classify.accuracy(classifier, test_set))
```

```python
# apply the trained classifier on the current data
tsuki_final['feature'] = tsuki_final['text'].apply(dialogue_act_features)
def act_classify(feature):
    return classifier.classify(feature)
tsuki_final['polite_5'] = tsuki_final['feature'].apply(act_classify)
```

```python
# save as csv
tsuki_final[['language','emoji','text','is_emoji',
       'num_emoji','polite_5']].to_csv('tsuki_polite5.csv')
```
